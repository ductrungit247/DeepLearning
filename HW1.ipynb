```python
def gradient_descent(initial, learning_rate, iterations):
    def f(x):
        return x**2

    def gradient(x):
        return 2*x

    x = initial
    x_history = [x]

    for i in range(iterations):
        x = x - learning_rate * gradient(x)
        x_history.append(x)

    x_values = np.linspace(-12, 12, 400)
    y_values = f(x_values)

    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label='f(x) = x^2')
    plt.scatter(x_history, [f(x) for x in x_history], c='red', marker='o', label='Gradient Descent Updates')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.title(f'Gradient Descent Optimization for f(x) = x^2\nInitial: {initial}, Learning Rate: {learning_rate}, Iterations: {iterations}')
    plt.grid(True)
    plt.show()

    return x
```

```python
optimized_x = gradient_descent(initial=10, learning_rate=0.2, iterations=10)
print(f'Optimized x: {optimized_x}')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def linear_regression(data, learning_rate, num_iterations):
    x = data[:, 0].reshape(-1, 1)
    y = data[:, 1].reshape(-1, 1)

    # Add a column of ones to the input features for the bias term
    x = np.hstack((np.ones((x.shape[0], 1)), x))

    # Initialize the weights
    w = np.array([0., 1.]).reshape(-1, 1)
    cost_history = []

    for i in range(num_iterations):
        # Calculate the predictions
        predictions = np.dot(x, w)

        # Calculate the error
        error = predictions - y

        # Calculate the cost (mean squared error)
        cost = 0.5 * np.sum(error ** 2)
        cost_history.append(cost)

        # Update the weights using gradient descent
        gradient = np.dot(x.T, error)
        w -= learning_rate * gradient

    # Plot the cost history
    plt.figure()
    plt.plot(range(num_iterations), cost_history)
    plt.xlabel('Iteration')
    plt.ylabel('Cost')
    plt.title('Cost vs. Iteration')

    # Plot the regression line
    predict = np.dot(x, w)
    plt.figure()
    plt.scatter(x[:, 1], y)
    plt.xlabel('mét vuông')
    plt.ylabel('giá')
    plt.plot(x[:, 1], predict, 'r')
    plt.show()

    return w
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def linear_regression(data, learning_rate, num_iterations):
    x = data[:, 0].reshape(-1, 1)
    y = data[:, 1].reshape(-1, 1)

    # Add a column of ones to the input features for the bias term
    x = np.hstack((np.ones((x.shape[0], 1)), x))

    # Initialize the weights
    w = np.array([0., 1.]).reshape(-1, 1)
    cost_history = []

    for i in range(num_iterations):
        # Calculate the predictions
        predictions = np.dot(x, w)

        # Calculate the error
        error = predictions - y

        # Calculate the cost (mean squared error)
        cost = 0.5 * np.sum(error ** 2)
        cost_history.append(cost)

        # Update the weights using gradient descent
        gradient = np.dot(x.T, error)
        w -= learning_rate * gradient

    # Plot the loss function
    plt.figure()
    plt.plot(range(num_iterations), cost_history)
    plt.xlabel('Iteration')
    plt.ylabel('Cost')
    plt.title('Cost vs. Iteration')

    # Plot the regression line
    predict = np.dot(x, w)
    plt.figure()
    plt.scatter(x[:, 1], y)
    plt.xlabel('mét vuông')
    plt.ylabel('giá')
    plt.plot(x[:, 1], predict, 'r')
    plt.show()

    return w, cost_history
```

```python

# Load the dataset
data = pd.read_csv('data_linear.csv').values

# Specify the learning rate and number of iterations
learning_rate = 0.000001
num_iterations = 100

# Perform Linear Regression and get the weights and cost history
weights, cost_history = linear_regression(data, learning_rate, num_iterations)

# Predict the price for 50m^2
x1 = 50
y1 = weights[0] + weights[1] * x1
print('Giá nhà cho 50m^2 là:', y1[0])
```
